 Tout d'abord, on importe deux classes importantes de scikit-learn : TfidfVectorizer pour convertir le texte en une représentation numérique appelée matrice TF-IDF, et cosine_similarity pour calculer la similarité cosinus entre les film. Ensuite, il utilise TfidfVectorizer pour créer une matrice TF-IDF à partir de la colonne "description" du sous-ensemble de données. En utilisant cette matrice, il calcule la similarité cosinus entre les films et identifie les contenus similaires en parcourant chaque contenu pour trouver les contenus les plus similaires en termes de description. Enfin, il imprime les contenus similaires pour chaque contenu du sous-ensemble de données, en excluant le contenu lui-même, avec leur titre et leur score de similarité cosinus.

```python
#Question 5:Identifier des contenus similaires en faisant correspondre des caractéristiques textuelles

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Sélectionner les 100 premières lignes
df_subset = df.head(100)

# Créer une instance de TfidfVectorizer pour convertir le texte en une matrice TF-IDF
tfidf_vectorizer = TfidfVectorizer(stop_words='english')

# Calculer la matrice TF-IDF pour le sous-ensemble de données
tfidf_matrix = tfidf_vectorizer.fit_transform(df_subset['description'])

# Calculer la similarité cosinus entre les documents
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Parcourir les contenus pour identifier les contenus similaires
for idx, row in df_subset.iterrows():
    similar_indices = cosine_sim[idx].argsort()[:-6:-1] # Les 5 premiers indices des contenus similaires (à l'exclusion de lui-même)
    similar_items = [(cosine_sim[idx][i], df_subset['title'].iloc[i]) for i in similar_indices]

    # Affichage des résultats
    print(f"Contenu similaire à '{row['title']}':")
    for i, (score, title) in enumerate(similar_items[1:]): # Commencer à 1 pour exclure lui-même
        print(f"{i+1}. {title} (Similarité: {score})")
    print("\n")
```

 Apres avoir choisit le style et la palette qui sera utilisé on souhaite comparer le nombre de series avec celui des films pour les 4 ans derniers  de cette database et puisque la derniere année ou un film ou serie a été ajouter est 2021 dans il fallait retourner six ans en arriere c'est pouquoi ona determiné la date il y avait six ans à partir d'aujourd'hui et on l'a stocker dans la variable four_years_ago. par la suite on a filtrer les données en comparant la valeur de la colonne  "released_year" a celle de four_years_ago et on stocke ceux qui sont superieur a  four_years_ago dans la variable recent_data , on regroupe les données par 'release_year' et le 'type' et on affiche les resultats dans un graphique à barres empilées

```python
#Question 6: Netflix se concentre-t-il davantage sur les séries télévisées que sur les films ces dernières années ?

from datetime import datetime, timedelta

# Set Seaborn style and palette
sns.set_style("whitegrid")
colors = sns.color_palette("Set1")

# Déterminer la date d'il y a six ans à partir d'aujourd'hui
four_years_ago = datetime.now() - timedelta(days=6*365)

# Filtrer les données pour les six dernières années
recent_data = df[df['release_year'] >= four_years_ago.year]

# Comptage par type et année (yearly_counts)
type_counts = recent_data.groupby(['release_year', 'type']).size().unstack().fillna(0)


# Création du graphique à barres empilées
type_counts.plot(kind='bar', stacked=True, color=colors)

plt.title('Évolution de la production de films et séries Netflix par année (dernières 6 années)')
plt.xlabel('Année')
plt.ylabel('Nombre de titres')

plt.show()

```

On Divise les films et les séries télévisées en groupes en fonction les catégories : "little_kids", "older_kids", "mature", etc.,
puis on supprime les lignes avec des valeurs manquante de la colonne " rating "
les classifications de contenu sont regroupées en listes selon les groupes d'âge, tels que "little_kids", "older_kids", "teens" et "mature". Les données sont ensuite regroupées par groupe par type (Movie ou TV SHOW).

```python
# Division des films et les séries télévisées en groupes en fonction les catégories : "little_kids", "older_kids", "mature", etc.,
little_kids = ['G', 'TV-Y', 'TV-G']
older_kids = ['PG', 'TV-Y7', 'TV-Y7-FV', 'TV-PG']
teens = ['PG-13', 'TV-14']
mature = ['R', 'NC-17', 'TV-MA']
rating_list = [little_kids, older_kids, teens, mature]
rating_names = ['Little kids', 'Older kids', 'Teens', 'Mature']

# Filter out rows with missing ratings
df_filtered = df[~df['rating'].isna()]

# Categorize content into age groups
age_group = np.select([df_filtered['rating'].isin(categ) for categ in rating_list], rating_names, 'Other')

# Group data by age group and type (movie/tv show)
df_grouped = df_filtered.groupby([age_group, 'type']).size().unstack(fill_value=0)

# Create a pie chart to visualize content distribution by age group
fig, ax = plt.subplots(figsize=(15, 6))
df_grouped.plot.pie(subplots=True, ax=ax, legend=True, autopct='%1.1f%%',
                    textprops={'fontsize': 12, 'color': 'white'},
                    title='Content Distribution by Age Group',
                    colors=sns.color_palette("Blues", n_colors=5))

# Show the plot
plt.show()
```

Pour recuperer Top 10 Nations for TV Shows on filtre le dataframe df par la colonne "type" egal a TV Show  on utilise la methode value_counts qui retourne le nombre de fois où chaque country apparaît dans la colonne "country" du DataFrame df suivie de head( 10) pour obtenir les dix premières country avec le plus grand nombre de TV Show .par la suite on affiche les resultats obtenus sous forme d'un diagramme de bar ou en utilisant simplement le print().

```python
#Top Nations for TV Shows

# Filter DataFrame for TV shows
tvshow_df = df[df['type'] == 'TV Show']

# Count the number of TV shows per country and select the top 10
country_counts = tvshow_df['country'].value_counts().head(10)

# Create a bar plot
plt.figure(figsize=(12, 6))
ax = sns.barplot(x=country_counts.index, y=country_counts.values, hue=country_counts.index, palette='cool', legend=False)

# Set plot title and axis labels
plt.title('Top 10 Countries with the Most TV Shows', fontsize=16)
plt.xlabel('Countries', fontsize=14)
plt.ylabel('Number of TV Show Entries', fontsize=14)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Add data labels to each bar
for i, count in enumerate(country_counts):
    plt.text(i, count, str(count), ha='center', va='bottom')

# Show the plot
plt.tight_layout()
plt.show()
```

On compte le nombre d'occurrences de chaque année de sortie dans la colonne "release_year" du DataFrame df. Cela signifie qu'elle compte combien de fois chaque année apparaît dans la colonne .'sort_index(): Cette méthode trie les résultats par index, c'est-à-dire par année de sortie. Cela garantit que les années sont triées dans l'ordre croissant.yearly_counts: Cette variable contient maintenant le nombre de films et de séries télévisées sortis chaque année, avec les années triées dans l'ordre croissant. puis on affiche les informations sur le graphique de l'évolution du nombre de sorties de contenus par année.

```python

#Évolution du nombre de sorties de contenus Netflix par année

# Agréger les données par année de sortie
yearly_counts = df['release_year'].value_counts().sort_index()

# Tracer le graphique de l'évolution du nombre de sorties de contenus par année
plt.figure(figsize=(10, 6))
plt.plot(yearly_counts.index, yearly_counts.values, marker='o', linestyle='-')
plt.title('Évolution du nombre de sorties de contenus Netflix par année')
plt.xlabel('Année de sortie')
plt.ylabel('Nombre de sorties de contenus')
plt.grid(True)
plt.show()
```

on premier temp on recupère les lignes ou 'release_year' est 2021 et le 'type' est Movie ,puis on crée une nouvelle colonne  nommé 'month_added' on stock le premier element de chaque case de la collone "date_added" qui est le mois en utilisant le separateur espace et puis on calcul le nombre de movies ajouté pour chaque mois de l'année 2021 et on finit par les afficher soit sous forme de diagramme avec bar ou en utilisant print()

```python
#Number of Movie Releases per Month in 2021

# 1. Filter movies with release year 2021
movies_2021 = df[(df['release_year'] == 2021) & (df['type'] == 'Movie')]

# 2. Split the "date_added" column and extract the month
movies_2021['month_added'] = movies_2021['date_added'].apply(lambda x: x.split()[0])

# 3. Count the number of movies added for each month
movies_added_per_month = movies_2021['month_added'].value_counts().sort_index()

# 4. Display the result
print(movies_added_per_month)

plt.figure(figsize=(10, 6))
movies_added_per_month.plot(kind='bar', color='skyblue')
plt.title('Number of Movie Releases per Month in 2021')
plt.xlabel('Month')
plt.ylabel('Number of Movie Releases')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
